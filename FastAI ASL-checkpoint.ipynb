```python
# Fastai Libraries and these libraries are being used to be readily avaialbe and fast productive
import fastai
from fastai import *
from fastai.vision import *
```

```python
help(untar_data)
```

```python
# storing the data from the source into a variable
path = pathlib.PurePosixPath('E:\Thesis\dataset')
path
```

```python
# bs to define the batch size to define number of training samples used in one iteration
bs = 8
```

```python
# Data is stored in 29 different folders name and To extract and classify the images data into right categories, Fastai library 
# gives unique function name ImageDataBunch where it automatically assigns the labels to the data as they are stored in their 
# respective folders.
# Normalizing the data using Pretrained Imagenet model.

data = ImageDataBunch.from_folder(path=path,bs=bs,ds_tfms=get_transforms(),size=200,valid_pct=0.2).normalize(imagenet_stats)
```

```python
# To Print all the classes respective to the data stored in the folders.

print(data.classes)
data.c
```

```python
# To show the Image data present in the folders.
data.show_batch(rows=3)
```

```python
# Creating a CNN model using a pretrained (resnet34) model which has Millions of images trained information stored in it.

learn = cnn_learner(data,models.resnet34,metrics=accuracy)
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
# This code helps to understand the number of hidden layers defined in the CNN model. And also helps to understand the types 
# of hyperparameters used in it for effiecient modelling.
learn.model
```

```python
# fit_one_cycle is used to train the model with cylical learning rates based on the iteration defined.
# And also it helps to interpret the train_loss, valid_loss, accuracy, time.

learn.fit_one_cycle(2)
```

```python
learn.show_results()
```

```python
# This line of code is used to store the train model as a checkpoint

learn.save('stage-1')
```

```python
# These lines of code helps to understand the losses, helps to create confusion matrix and which is used to visualize
# incorrect images predicted.

interp = ClassificationInterpretation.from_learner(learn)

losses,idxs = interp.top_losses()

len(data.valid_ds)==len(losses)==len(idxs)
```

```python
# This line of code prints the prediction/ loss/ probability values

interp.plot_top_losses(9,figsize=(15,11))
```

```python
# To identify the top losses between different categories
doc(interp.plot_top_losses)
```

```python
# To plot the confusion matrix and interpret the losses and incorrect category predicted.
interp.plot_confusion_matrix(figsize=(12,12), dpi=60)
```

```python
# This line of code gives the text interpretion of confusion matrix.
interp.most_confused(min_val=2)
```

```python
# Model Fine-tuning
```

```python
# This line of code is used to train the model from the scratch and all the layers are trained as per the loss function.
learn.unfreeze()
```

```python
# Re training the model on unfreezed weights.
learn.fit_one_cycle(1)
```

```python
# Helps to find the best learning rate by plotting a graph.
learn.lr_find()
```

```python
# Displays the plot
learn.recorder.plot()
```

```python
# Retraining the model based on the best learning rate identified using the graph
# Where max_lr=slice() is used to define the range of learning rate.

learn.unfreeze()
learn.fit_one_cycle(2, max_lr=slice(1e-2,1e-1))
```

```python
import torch
import torchvision.models
import hiddenlayer as hl
```

```python
# Resnet101
model = torchvision.models.resnet34()

# Rather than using the default transforms, build custom ones to group
# nodes of residual and bottleneck blocks.
transforms = [
    # Fold Conv, BN, RELU layers into one
    hl.transforms.Fold("Conv > BatchNorm > Relu", "ConvBnRelu"),
    # Fold Conv, BN layers together
    hl.transforms.Fold("Conv > BatchNorm", "ConvBn"),
    # Fold bottleneck blocks
    hl.transforms.Fold("""
        ((ConvBnRelu > ConvBnRelu > ConvBn) | ConvBn) > Add > Relu
        """, "BottleneckBlock", "Bottleneck Block"),
    # Fold residual blocks
    hl.transforms.Fold("""ConvBnRelu > ConvBnRelu > ConvBn > Add > Relu""",
                       "ResBlock", "Residual Block"),
    # Fold repeated blocks
    hl.transforms.FoldDuplicates(),
]

# Display graph using the transforms above
hl.build_graph(model, torch.zeros([1, 3, 224, 224]), transforms=transforms)
```
